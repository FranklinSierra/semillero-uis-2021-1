## <font color='green'>What is computer vision? </font>

1.jpg


* Computer vision is an <font color ='red'>**interdisciplinary scientific field**</font> that deals with how computers can gain high-level understanding from digital images or videos.

* Set of <font color='red'>**computer methods**</font> and techniques aimed at interpreting images and videos <font color='red'>**to replicate human visual abilities.**</font>







## <font color='green'>What is an image?</font>

2.gif

* Digital images are numerical representations (matrices) with three channels, <font color='red'>Red</font>, <font color='green'>Green</font> and <font color='blue'>Blue</font> (RGB)
* The size of the image is determined by the number of pixels. for example, an image with a size of $3100\times3100$ is $3100$ pixels high and $3100$ pixels wide
*    Each pixel represent a color value intensities between $[0-255] = 2^8$. In color, each matrix reprent the quantity of color at each channel

```
from PIL import Image
import matplotlib.pyplot as plt
#from skimage.transform import resize
#from skimage.color import rgb2hed
```

**Load images**


```
path1 = "/content/drive/MyDrive/semillero Bivl2ab 2021-1/data/serrated_NBI.png"
img1 = Image.open(path1)
print("image shape: ", img1.size)
```

**Print images**

```
plt.figure(figsize=(8,8)) # creates a figure 8 inches by 8 inches
plt.imshow(img1)
print("axis range 3100 pixels x 3100 pixels")
```

```
x, y = 500, 200
plt.figure(figsize=(8,8))
plt.scatter(x,y, marker="x", s=200, linewidth=5, c="red")
plt.imshow(img1)
print("marker position: ", x, y)
print("pixel value:", img1[y,x,:] )
```

```
plt.figure(figsize=(16,16))
print("image channels")
plt.subplot(131),plt.imshow(img1[:,:,0]),plt.title("Red")
plt.subplot(132),plt.imshow(img1[:,:,1]),plt.title("Green")
plt.subplot(133),plt.imshow(img1[:,:,2]),plt.title("Blue")
```
