### **Loading data on the fly**

```
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
     rotation_range=40,
     horizontal_flip=True,
     fill_mode='reflect'
)
```

```
path = train_df.iloc[idx]["path"]
img = imageio.imread(path)
img = img.reshape((1,) + img.shape)
imgtransform = datagen.flow(img, batch_size=1).next()
```

```
plt.figure(figsize=(16,16))
plt.subplot(121),plt.imshow(img.reshape((576,768,3))),plt.title("Original")
plt.subplot(122),plt.imshow(imgtransform.astype(np.uint8).reshape((576,768,3))),plt.title("Transform")
```
